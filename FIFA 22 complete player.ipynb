## Insight 2

```python
#Grouping the dataframe according to league name by the value of the player
dfbar_plot = df.groupby("league_name")["value_eur"].sum().sort_values(ascending=False).head(5)
dfbar_plot = dfbar_plot.reset_index()
sns.barplot(x="league_name", y="value_eur",data=dfbar_plot)
```

#### Through this analysis we can find that English premier league has the more valuable players than any other league
